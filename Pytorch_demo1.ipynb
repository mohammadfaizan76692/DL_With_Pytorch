### Reshaping , Stacking , Squeezing and Unsqueezing

*   Reshape :  reshape a tensor into a defined shape

*  View :  Return a view of an input tensor of certain shape but    keep the same memory as the original tensor.
*   Stacking : combine multiple tensors on top of each  other (vstack) or side by side (hstack)


*   squeeze : removes all '1' dimensions from a tensor
*   Unsqueeze : add a 1 dimension to a target tensor
*   Permute : Return a view of the input with dimensions permuted(swapped) in a certain way



```python
import torch
x = torch.arange(9,18.)
x,x.shape
```

```python
# Add a extra dimension
# size should be same
```

```python
x.shape
```

```python
## view
# changing z changes x( because a view of tensor shares the same as the original tensor)
x_reshape = x.reshape(9,1)
z = x.view(1,9)
z
```

```python
z[:,0] = 5
```

```python
x
```

```python
# stack tensors on top of each other
# dim =0 vertically
# dim=1 horizontally
x_stacked = torch.stack([x, x, x, x, x],dim=0)
```

```python
x_stacked
```

```python
# torch.squeeze() -> removes all the single dimension from a target tensor
```

```python
x_reshape.shape
```

```python
squeezed_x_reshape = torch.squeeze(x_reshape)
```

```python
print(f"Orginal {x_reshape} shape -> {x_reshape.shape}")
print(f"Squeezed {squeezed_x_reshape} shape -> {squeezed_x_reshape.shape}")
```

```python
# unsqueeze() adding single dimension
unsqueezing = torch.unsqueeze(squeezed_x_reshape,dim =0)
print(f"Squeezed {squeezed_x_reshape} shape -> {squeezed_x_reshape.shape}")
print(f"UnSqueezed {unsqueezing} shape -> {unsqueezing.shape}")
```

```python
# torch.permute  -> rearranges the dimensions of a tareget tensor in a specified order it is aslo a view share the same memory
image = torch.rand(size = (224,224,3))

# permute the original tensor to rearrange the axis (or dim) order

permuted_image = image.permute(dims = (2,0,1))
print(f"image shape --> {image.shape}")
print(f"permuted image shape --> {permuted_image.shape}")
```

### Reproducibility (trying to take out random out of randoms)
In short how neural networks learns: start with random numbers -> tensor operations -> update random number to try and make them better representation of the data -> again  -> again -> again

To reduce the randomness in neural networks and pytorch comes the concept of a ** Random seed **.

Essentially what the random seed does is "flavour the randomness"




```python
import torch
```

```python

torch.manual_seed(42)
x = torch.randn(3,3)
torch.manual_seed(42)
t = torch.randn(3,3)
x==t
```
